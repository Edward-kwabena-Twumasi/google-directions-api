Duraction and trafficaDuration fields are mixed with second values and minutes. Convert seconds to minutes

```python
def convert_to_minutes(secs):
    if 'mins' in str(secs):
        return secs
    minutes = int(secs) // 60
    seconds_remainder = int(secs) % 60
    return f"{minutes} mins {seconds_remainder} secs"
```

```python
new_traffic["duration_min"] = new_traffic["duration"].apply(convert_to_minutes)
```

```python
new_traffic.head()
```

```python
id_groups = new_traffic.groupby("requestId")
```

```python
id_groups['duration_min'].unique()
```

```python
id_groups['duration_min'].nunique()
```

```python
new_traffic["t_duration_min"] = new_traffic["trafficDuration"].apply(convert_to_minutes)
```
